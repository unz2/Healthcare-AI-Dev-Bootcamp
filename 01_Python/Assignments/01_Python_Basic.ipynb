# 과제2. BMI 계산기

```python
height = float(input('키를 입력하세요 : '))
weight = float(input('몸무게를 입력하세요 : '))
bmi = round(weight / ((height/100)**2),1)

if bmi < 18.5:
    print(f'당신의 bmi는 {bmi}, 저체중 입니다.')
elif bmi < 23:
    print(f'당신의 bmi는 {bmi}, 정상 입니다.')
elif bmi < 25:
    print(f'당신의 bmi는 {bmi}, 과체중 입니다.')
elif bmi <= 30:
    print(f'당신의 bmi는 {bmi}, 비만 입니다.')
else:
    print(f'당신의 bmi는 {bmi}, 고도비만 입니다.')
```
