```python
patients = []

print ('ğŸ“‹ í™˜ì ì •ë³´ ê´€ë¦¬ ì½˜ì†”')

while True:    
    print ('==========================================')
    print ('1ï¸âƒ£ í™˜ì ì •ë³´ ì…ë ¥ 2ï¸âƒ£ í™˜ì ëª©ë¡ ë³´ê¸° 3ï¸âƒ£ í™˜ì ì´ë¦„ ê²€ìƒ‰')
    print ('4ï¸âƒ£ í™˜ì ì •ë³´ ìˆ˜ì • 5ï¸âƒ£ í™˜ì ì •ë³´ ì‚­ì œ 6ï¸âƒ£ í†µê³„')
    print ('7ï¸âƒ£ ë°ì´í„° ì´ˆê¸°í™”  8ï¸âƒ£ ì¢…ë£Œ')
    cmd = int(input('âœï¸ ë©”ë‰´ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš” :'))
    print ('==========================================')
    

    # 1. í™˜ìì •ë³´ ì…ë ¥ â†’ dict ë§Œë“¤ì–´ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    if cmd == 1:
        print ('1ï¸âƒ£ í™˜ì ì •ë³´ ì…ë ¥')
        name = input('ì´ë¦„ :').strip()
        
        try:
            age = int(input('ë‚˜ì´ :'))
        except ValueError:
            print('âŒ ë‚˜ì´ëŠ” ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”')
            continue
            
        gender = input('ì„±ë³„(M/F) :').strip().upper()

        if not all([name, age, gender]):
            print('âŒ í™˜ì ì •ë³´ê°€ ëˆ„ë½ë˜ì—ˆìŠµë‹ˆë‹¤.')
            continue

        if gender not in ['M','F']:
            print('âŒ ì„±ë³„ì„ M ë˜ëŠ” Fë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
            continue

        patients.append({
            'name':name,
            'age':age,
            'gender':gender,
        })

    
    # 2. í™˜ìëª©ë¡ ë³´ê¸° â†’ forì™€ enumerate(start=1)ë¡œ ìˆœë²ˆê³¼ í•¨ê»˜ ì¶œë ¥
    elif cmd == 2:
        print ('2ï¸âƒ£ í™˜ì ëª©ë¡ ë³´ê¸°')
        if not patients:
            print('ğŸ™… ë“±ë¡ëœ í™˜ìê°€ ì—†ìŠµë‹ˆë‹¤.') 
        else:
            for i, patient in enumerate(patients, start=1):
                print(f'{i}. {patient.get('name')}({patient.get('age')}ì„¸, {patient.get('gender')}) {patient.get('memo','')}')
            

    # 3. ì´ë¦„ìœ¼ë¡œ ê²€ìƒ‰ (ë¶€ë¶„ì¼ì¹˜) â†’ for + ì¡°ê±´
    elif cmd == 3:
        print ('3ï¸âƒ£ í™˜ì ì´ë¦„ ê²€ìƒ‰')
        search_name3 = input('â–¶ï¸ ê²€ìƒ‰í•  í™˜ì ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš”.')
        matches3 = [f'{patient.get('name')}({patient.get('age')}ì„¸, {patient.get('gender')})'
                   for patient in patients
                   if search_name3 in patient.get('name')]
        
        if not matches3:
            print('ğŸ™… ê²€ìƒ‰ ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.')
        else:
            for i, match in enumerate(matches3, start=1):
                print(f'{i}. {match}')
        

    # 4. í™˜ìì •ë³´ ìˆ˜ì • (ë©”ëª¨ë§Œ ê°„ë‹¨íˆ) â†’ ì„ íƒí•œ í™˜ìì˜ memo ì—…ë°ì´íŠ¸
    elif cmd == 4:
        print ('4ï¸âƒ£ í™˜ì ì •ë³´ ìˆ˜ì •')
        select_name4 = input('â–¶ï¸ ìˆ˜ì •í•  í™˜ì ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš” :')
        matches4 = [patient.get('name') for patient in patients if select_name4 == patient.get('name')]

        if not matches4:
            print('ğŸ™… ê²€ìƒ‰ ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.')
        else:
            for patient in patients:
                if patient.get('name') == select_name4:
                    update_memo = input('â–¶ï¸ ì¶”ê°€/ìˆ˜ì •í•  ë©”ëª¨ ë‚´ìš©ì„ ì…ë ¥í•´ì£¼ì„¸ìš” :')
                    patient.update({
                        'memo':update_memo
                    })

    
    # 5. í™˜ì ì‚­ì œ â†’ ì„ íƒí•œ ë²ˆí˜¸ì˜ í™˜ì ì œê±°
    elif cmd == 5:
        print ('5ï¸âƒ£ í™˜ì ì •ë³´ ì‚­ì œ')
        print('í˜„ì¬ í™˜ì ëª©ë¡')
        for i, patient in enumerate(patients, start=1):
            print(f'{i}. {patient.get('name')}({patient.get('age')}ì„¸, {patient.get('gender')}) {patient.get('memo','')}')

        remove_idx = int(input('â–¶ï¸ ì‚­ì œí•  í™˜ì ë²ˆí˜¸ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš” :'))
        patients.pop(remove_idx - 1)

    
    # 6. ê°„ë‹¨í†µê³„ â†’ ì´ í™˜ì ìˆ˜ (len), í‰ê·  ë‚˜ì´ (sum / len, í™˜ì ì—†ìœ¼ë©´ 0), ì„±ë³„ ë¶„í¬ (count)
    elif cmd == 6:
        print ('6ï¸âƒ£ í†µê³„')
        gender_list = ['M' if patient.get('gender') == 'M' else 'F' for patient in patients]

        if len(patients) == 0:
            average_age = 0
        else:
            total_age = 0
            for patient in patients:
                total_age += patient.get('age')
            average_age = round(total_age / len(patients),1)
        
        print(f'ì´ í™˜ì ìˆ˜ : {len(patients)}ëª…')
        print(f'í™˜ì í‰ê·  ë‚˜ì´ : {average_age}ì„¸')
        print(f'ì„±ë³„ë¶„í¬(ë‚¨ì) : {gender_list.count('M')}ëª…')
        print(f'ì„±ë³„ë¶„í¬(ì—¬ì) : {gender_list.count('F')}ëª…')


    # 7. ë°ì´í„° ì´ˆê¸°í™”
    elif cmd == 7:
        print ('7ï¸âƒ£ ë°ì´í„° ì´ˆê¸°í™”')
        confirm = input('â–¶ï¸ ë°ì´í„°ë¥¼ ì´ˆê¸°í™” í•˜ì‹œê² ìŠµë‹ˆê¹Œ?(Y/N) :').strip().upper()
        if confirm == 'Y':
            patients.clear()
            print('âœ… ë°ì´í„° ì´ˆê¸°í™” ì™„ë£Œ')
        else:
            print('âŒ ë°ì´í„° ì´ˆê¸°í™” ì·¨ì†Œ')
            continue

    
    # 8. ì¢…ë£Œ â†’ break
    elif cmd == 8:
        print ('8ï¸âƒ£ ì¢…ë£Œ')
        print('ğŸ“‹ í™˜ì ì •ë³´ ê´€ë¦¬ ì½˜ì†”ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.')
        break;

    
    elif cmd not in range(1,9):
        print('âš ï¸ 1 ~ 8ë²ˆ ë©”ë‰´ ë²ˆí˜¸ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš” âš ï¸')
        continue
        
```
