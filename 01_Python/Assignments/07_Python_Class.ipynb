## 1) ë°ì½”ë ˆì´í„° ì—†ì´ ë™ì‘ ì´í•´í•˜ê¸° (ë¬¸ë²• í¼ì¹˜ê¸°)

```python

def log_decorator(func):
    def wrapper():
        print("[LOG] í•¨ìˆ˜ ì‹œì‘")
        func()
        print("[LOG] í•¨ìˆ˜ ë")
    return wrapper

def greet():
    print("ì•ˆë…•í•˜ì„¸ìš”!")

# TODO: ë°ì½”ë ˆì´í„° ë¬¸ë²•(@...) ëŒ€ì‹  í•¨ìˆ˜ ëŒ€ì…ìœ¼ë¡œ ë™ì¼ íš¨ê³¼ ë§Œë“¤ê¸°
# ì˜ˆ: greet = ____________  # <-- ë¹ˆì¹¸ ì±„ìš°ê¸°ìš© ì˜ˆì‹œ: í’€ì´ ì‹œ ì´ ì¤„ë§Œ ë‚¨ê¸°ê³  ìœ„ëŠ” ì£¼ì„/ìˆ˜ì •í•´ë„ ë¨

greet = log_decorator(greet)

# í…ŒìŠ¤íŠ¸ (ì‹œê° í™•ì¸)
greet()
```


## 3) ìº¡ìŠí™” + `@property`

ìš”êµ¬ì‚¬í•­
- MapleCharacter í´ë˜ìŠ¤ ë§Œë“¤ê¸°
- __hpëŠ” ì™¸ë¶€ì—ì„œ ì§ì ‘ ì ‘ê·¼ ë¶ˆê°€
- get_hp(), set_hp() ë©”ì„œë“œë¡œ HPë¥¼ ì•ˆì „í•˜ê²Œ ê´€ë¦¬
- @property / @hp.setter ë¡œ ê°™ì€ ê¸°ëŠ¥ì„ ì†ì„± ì ‘ê·¼ í˜•íƒœë¡œë„ êµ¬í˜„

```python

class MapleCharacter:
    def __init__(self, name):
        self.name = name
        self.__hp = 100   # private ë³€ìˆ˜

    # âœ… 1ï¸âƒ£ getter
    def get_hp(self):
        return self.__hp

    # âœ… 2ï¸âƒ£ setter
    def set_hp(self, value):
        if value < 0:
            self.__hp = 0
        elif value > 1000:
            self.__hp = 1000
        else:
            self.__hp = value
            
    # âœ… 3ï¸âƒ£ property getter
    @property
    def hp(self):
        return self.__hp

    # âœ… 4ï¸âƒ£ property setter
    @hp.setter
    def hp(self, value):
        if value < 0:
            self.__hp = 0
        elif value > 1000:
            self.__hp = 1000
        else:
            self.__hp = value

# ğŸ§ª í…ŒìŠ¤íŠ¸ (assert ìë™ ì±„ì )
m = MapleCharacter("ë£¨ì¹´")

# â‘  getter/setter í…ŒìŠ¤íŠ¸
m.set_hp(1200)
assert m.get_hp() == 1000, "âŒ set_hp(1200) â†’ HPëŠ” 1000ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

m.set_hp(-10)
assert m.get_hp() == 0, "âŒ set_hp(-10) â†’ HPëŠ” 0ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

m.set_hp(500)
assert m.get_hp() == 500, "âŒ set_hp(500) â†’ HPëŠ” 500ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

# â‘¡ property í…ŒìŠ¤íŠ¸
m.hp = 2000
assert m.hp == 1000, "âŒ m.hp=2000 â†’ HPëŠ” 1000ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

m.hp = -50
assert m.hp == 0, "âŒ m.hp=-50 â†’ HPëŠ” 0ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

m.hp = 750
assert m.hp == 750, "âŒ m.hp=750 â†’ HPëŠ” 750ì´ì–´ì•¼ í•©ë‹ˆë‹¤"

print("âœ… 3) í†µê³¼")
```
