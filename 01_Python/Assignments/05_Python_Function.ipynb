## 문제 1) for문 기본 — 이름 인사하기
리스트의 각 이름에 대해 "안녕, 이름" 형태의 문장을 만들어 `outputs`에 담으세요.

```python
names = [
    "철수",
    "영희",
    "민수",
]

outputs = []

# TODO: 기본 for문 — 순서대로 순회
for name in names:
    outputs.append(f"안녕, {name}")

assert outputs == ["안녕, 철수", "안녕, 영희", "안녕, 민수"]
print("✅ 문제 1 통과")
```

## 문제 2) 이터러블 — 문자열 순회
`text`의 각 문자를 하나씩 꺼내 리스트로 만드세요.

```python
text = "Py"

# TODO: 문자열도 이터러블 — 한 글자씩 순회
chars = [t for t in text]

assert chars == ['P','y']
print("✅ 문제 2 통과")
```

## 문제 4) enumerate — 1부터 번호 붙이기
`enumerate`로 1부터 시작하는 `(번호, 값)` 리스트를 만드세요.

```python
items = [
    "a",
    "b",
    "c",
]

# TODO: enumerate(iterable, start=1)
indexed = [
    (i, v)
    for i, v in enumerate(items, start=1)
]

assert indexed == [(1,'a'), (2,'b'), (3,'c')]
print("✅ 문제 4 통과")
```

## 문제 5) zip — 두 리스트 동시 순회
이름과 점수를 **쌍**으로 묶어 `(이름, 점수)` 리스트를 만드세요.

```python
names = [
    "철수",
    "영희",
    "민수",
]
scores = [
    85,
    92,
    78,
]

# TODO: zip(names, scores)로 (이름, 점수) 튜플 생성
paired = [
    (n, s)
    for n, s in zip(names, scores)
]

assert paired == [("철수", 85), ("영희", 92), ("민수", 78)]
print("✅ 문제 5 통과")
```

## 문제 6) zip + enumerate — 가독성 좋게 쓰기
아래와 같이 **줄바꿈**으로 데이터 흐름을 보이게 하세요. 
zip으로 묶은 뒤, enumerate로 1부터 번호를 붙여 `(i, name, score)` 튜플을 만드세요.

```python
names = ["철수", "영희", "민수"]
scores = [85, 92, 78]

# TODO: zip -> enumerate(start=1)
triples = [
    (i, n, s)
    for i, (n, s) in enumerate(zip(names, scores), start=1)
]

assert triples == [
    (1, "철수", 85),
    (2, "영희", 92),
    (3, "민수", 78),
]
print("✅ 문제 6 통과")
```

## 문제 7) dict 이터러블 — 키로 순회
딕셔너리는 기본적으로 **키**를 순회합니다. `rows`에 `"key:value"` 문자열을 담으세요. (순서 무관)

```python
person = {
    "name": "철수",
    "age": 20,
}

rows = []

# TODO: for key in dict:
for key in person:
    rows.append(f"{key}:{person[key]}")

assert set(rows) == {"name:철수", "age:20"}
print("✅ 문제 7 통과")
```

## 문제 8) set 이터러블 — 순회하여 수집
집합은 순서가 없지만, **모든 원소를 한 번씩 순회**할 수 있습니다. `got`에 10,20,30을 모두 담으세요. (순서 무관)

```python
s = {10, 20, 30}
got = []

# TODO: for v in set:
for v in s:
    got.append(v)

assert set(got) == {10, 20, 30}
print("✅ 문제 8 통과")
```

## 문제 9) 중첩 for문 — 미니 구구단(2~3단, 1~3)
중첩 for문으로 `'2 x 1 = 2'` 형태 문자열을 만들어 리스트 `table`에 담으세요.

```python
table = []

# TODO: 바깥 단(i): range(2, 4), 안쪽 곱(j): range(1, 4)
for i in range(2, 4):
    for j in range(1, 4):
        table.append(f"{i} x {j} = {i*j}")

assert table[:3] == ['2 x 1 = 2','2 x 2 = 4','2 x 3 = 6']
assert table[-3:] == ['3 x 1 = 3','3 x 2 = 6','3 x 3 = 9']
print("✅ 문제 9 통과")
```

## 문제 10) while — 카운트다운
`x`가 0이 될 때까지 감소시키며 `[3,2,1]`을 `log`에 기록하세요.

```python
x = 3
log = []

# TODO: while x > 0:
while x > 0:
    log.append(x)
    x -= 1

assert log == [3, 2, 1] and x == 0
print("✅ 문제 10 통과")
```

## 문제 11) while + break — 1,2,3을 담고 탈출
무한 루프에서 1,2,3을 `out`에 담고 `break`로 탈출하세요. (출력 대신 리스트로 검증)

```python
out = []
x = 1

# TODO: while True + break 조건
while True:
    out.append(x)
    if x == 3:
        break
    x += 1

assert out == [1,2,3]
print("✅ 문제 11 통과")
```

## 문제 12) for + break — 5에서 멈추기
1부터 순회하며 5에서 멈추고 `seen == [1,2,3,4]`가 되게 하세요.

```python
seen = []

# TODO: range(1, 10) 순회 + if i == 5: break
for i in range(1, 10):
    if i == 5:
        break
    seen.append(i)

assert seen == [1,2,3,4]
print("✅ 문제 12 통과")
```

## 문제 13) for + continue — 홀수만 수집
1부터 5까지 순회하며 홀수만 `odds`에 담으세요(`[1,3,5]`).

```python
odds = []

# TODO: 짝수면 continue
for i in range(1, 6):
    if i % 2 == 0:
        continue
    odds.append(i)

assert odds == [1,3,5]
print("✅ 문제 13 통과")
```

## 문제 14) 리스트 컴프리헨션 — 제곱 리스트
`[1,2,3,4,5]`의 제곱 리스트를 만드세요.

```python
# TODO: [표현식 for 변수 in iterable]
squares = [
    x**2              # 힌트: x**2
    for x in range(1, 6)  # 힌트: range(1, 6)
]

assert squares == [1,4,9,16,25]
print("✅ 문제 14 통과")
```

## 문제 15) 리스트 컴프리헨션 + 조건 — 짝수만
`0..9`에서 짝수만 `evens`에 담으세요.

```python
# TODO: if 조건으로 필터링
evens = [
    x
    for x in range(10)
    if x % 2 == 0         # 힌트: x % 2 == 0
]

assert evens == [0,2,4,6,8]
print("✅ 문제 15 통과")
```

## 문제 16) 중첩 컴프리헨션 — 미니 구구단(2~3단, 1~3)
`['2x1=2', '2x2=4', ..., '3x3=9']` 형태를 만들어 `gugu`에 담으세요.

```python
gugu = [
    f"{i}x{j}={i*j}"
    for i in range(2, 4)   # 힌트: range(2, 4)
    for j in range(1, 4)   # 힌트: range(1, 4)
]

assert gugu == ['2x1=2','2x2=4','2x3=6','3x1=3','3x2=6','3x3=9']
print("✅ 문제 16 통과")
```

## 문제 17) 문자열 처리 — 대문자 리스트

## 문제 20) 실용 예제 — 성인 환자 이름만(컴프리헨션)
나이 20 이상 환자의 이름만 리스트 `adults`로 만드세요.

```python
patients = [
    {"name":"철수","age":25},
    {"name":"영희","age":30},
    {"name":"민수","age":17},
]

# TODO: 리스트 컴프리헨션 + 조건
adults = [
    p.get('name')
    for p in patients
    if p.get('age') >=20
]

assert adults == ["철수","영희"]
print("✅ 문제 20 통과")
```
