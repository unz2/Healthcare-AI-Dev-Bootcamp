## 문제 1 : global 키워드
- 전역 변수 `count`를 0으로 두고, 함수에서 1 증가시키기
- 빈칸을 채워 실행해보세요.

```python
count = 0

def increase():
    global count 
    count += 1

increase()
assert count == 1
print("✅ 문제 1 통과")
```

## 문제 2 : nonlocal 키워드
- 바깥 함수의 변수를 안쪽 함수에서 바꾸기
- `nonlocal`을 넣어 빈칸을 채우세요.

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x += 5
    inner()
    return x

result = outer()
assert result == 15
print("✅ 문제2 통과")
```

## 문제 3 : global + nonlocal 혼합
- 전역 변수 `score`와 바깥 변수 `bonus`를 동시에 바꾸어야 함
- 어디에 `global`, 어디에 `nonlocal`을 써야 할까요?

```python
score = 0
def outer():
    bonus = 1
    def inner():
        global score 
        nonlocal bonus
        score += 10
        bonus += 2
        return score, bonus
    return inner()

s, b = outer()
assert s == 10
assert b == 3
print("✅ 문제3 통과")
```
