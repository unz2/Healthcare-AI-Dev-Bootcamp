### **4. ë²¡í„°í™” ì—°ì‚° (Vectorized Operation)**

#### **ğŸ“œ ê°œë… ì„¤ëª…**
ë²¡í„°í™”ëŠ” NumPyì˜ ê°€ì¥ ê°•ë ¥í•œ ê¸°ëŠ¥ì…ë‹ˆë‹¤. `for` ë°˜ë³µë¬¸ ì—†ì´ ë°°ì—´ ì „ì²´ì— ëŒ€í•´ ì‚°ìˆ  ì—°ì‚°ì„ í•œ ë²ˆì— ì ìš©í•  ìˆ˜ ìˆê²Œ í•´ì¤ë‹ˆë‹¤. ì½”ë“œê°€ ë§¤ìš° ê°„ê²°í•´ì§€ê³  ì—°ì‚° ì†ë„ê°€ ë¹„ì•½ì ìœ¼ë¡œ ë¹¨ë¼ì§‘ë‹ˆë‹¤.

#### **Part 1: ìŠ¤ì¹¼ë¼-ë²¡í„° ì—°ì‚° (ë°°ì—´ê³¼ ìˆ«ì ê°„ì˜ ì—°ì‚°)**

##### **ğŸ› ï¸ ì‚¬ìš© ë°©ë²•**
NumPy ë°°ì—´ê³¼ ìˆ«ì(ìŠ¤ì¹¼ë¼) ì‚¬ì´ì— ì¼ë°˜ì ì¸ ì‚°ìˆ  ì—°ì‚°ì(`+`, `-`, `*`, `/`)ë¥¼ ì‚¬ìš©í•˜ë©´ ë©ë‹ˆë‹¤. ì—°ì‚°ì€ ë°°ì—´ì˜ ëª¨ë“  ì›ì†Œì— ìë™ìœ¼ë¡œ ì ìš©(ë¸Œë¡œë“œìºìŠ¤íŒ…)ë©ë‹ˆë‹¤.

```python
arr = np.array([1, 2, 3])
print("ë°°ì—´ì— 10 ë”í•˜ê¸°:", arr + 10)
print("ë°°ì—´ì— 2 ê³±í•˜ê¸°:", arr * 2)
```

##### **âœï¸ ì‹¤ìŠµ ë¬¸ì œ**
`patient_data` ë°°ì—´ì—ì„œ ëª¨ë“  í™˜ìì˜ **í‚¤(cm)ë¥¼ ë¯¸í„°(m) ë‹¨ìœ„ë¡œ ë³€í™˜**í•˜ì„¸ìš”.
1.  ë¨¼ì € ëª¨ë“  í™˜ìì˜ í‚¤ ë°ì´í„°(ì¸ë±ìŠ¤ 0ë²ˆ ì—´)ë¥¼ ì¶”ì¶œí•©ë‹ˆë‹¤.
2.  ì¶”ì¶œëœ ë°°ì—´ì„ 100ìœ¼ë¡œ ë‚˜ëˆ•ë‹ˆë‹¤.

```python
patient_height_m = patient_data[:, 0] / 100
print(patient_height_m)
```

#### **Part 2: ë²¡í„°-ë²¡í„° ì—°ì‚° (ë°°ì—´ê³¼ ë°°ì—´ ê°„ì˜ ì—°ì‚°)**
**ê°™ì€ í¬ê¸°**ë¥¼ ê°€ì§„ ë‘ ë°°ì—´ ì‚¬ì´ì˜ ì—°ì‚°ì€ ê°™ì€ ìœ„ì¹˜ì— ìˆëŠ” ì›ì†Œë¼ë¦¬(element-wise) ìˆ˜í–‰ë©ë‹ˆë‹¤.

##### **ğŸ› ï¸ ì‚¬ìš© ë°©ë²•**
í¬ê¸°ê°€ ë™ì¼í•œ ë‘ NumPy ë°°ì—´ ì‚¬ì´ì— ì‚°ìˆ  ì—°ì‚°ìë¥¼ ì‚¬ìš©í•˜ë©´, ê°™ì€ ìœ„ì¹˜ì˜ ì›ì†Œë¼ë¦¬ ì—°ì‚°ì´ ì´ë£¨ì–´ì§‘ë‹ˆë‹¤.

```python
midterm_scores = np.array([80, 75, 90])
final_scores = np.array([85, 95, 88])

# ê°™ì€ ì¸ë±ìŠ¤ì˜ ì›ì†Œë¼ë¦¬ ë”í•¨: 80+85, 75+95, 90+88
total_scores = midterm_scores + final_scores
print("í•©ì‚° ì´ì :", total_scores)
```

* **BMI ê³„ì‚° ê³µì‹**: `ì²´ì¤‘(kg) / (í‚¤(m) ** 2)`

```python
# ì£¼ì–´ì§„ ë°ì´í„°
weights_kg = np.array([72, 58, 85])
heights_m = np.array([1.75, 1.60, 1.82])
```

```python
bmi = weights_kg / heights_m **2
print(bmi)
```

### **5. ì¡°ê±´ í•„í„°ë§ (Boolean Indexing)**

#### **ğŸ“œ ê°œë… ì„¤ëª…**
ì¡°ê±´ í•„í„°ë§ì€ íŠ¹ì • ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ë°ì´í„°ë§Œ `True`ë¡œ í‘œì‹œí•˜ì—¬ ê³¨ë¼ë‚´ëŠ” ë§¤ìš° ìœ ìš©í•œ ê¸°ëŠ¥ì…ë‹ˆë‹¤. ë°ì´í„°ì˜ ë¶€ë¶„ì§‘í•©ì„ ì¶”ì¶œí•˜ì—¬ ë¶„ì„ì„ ì‹¬í™”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

#### **ğŸ› ï¸ ì‚¬ìš© ë°©ë²•**
1.  ë°°ì—´ì— ì¡°ê±´ì‹(ì˜ˆ: `ë°°ì—´ > 30`)ì„ ì ìš©í•˜ì—¬ `True`ì™€ `False`ë¡œ ì´ë£¨ì–´ì§„ **ë¶ˆë¦¬ì–¸ ë§ˆìŠ¤í¬(boolean mask)** ë°°ì—´ì„ ë§Œë“­ë‹ˆë‹¤.
2.  ìƒì„±ëœ ë§ˆìŠ¤í¬ë¥¼ ì›ë³¸ ë°°ì—´ì˜ ì¸ë±ìŠ¤ ìë¦¬ì— ë„£ìœ¼ë©´, ë§ˆìŠ¤í¬ ê°’ì´ `True`ì¸ ìœ„ì¹˜ì˜ ì›ì†Œë§Œ ì¶”ì¶œë©ë‹ˆë‹¤.

### **6. ê¸°ë³¸ í†µê³„ ê³„ì‚°**

#### **ğŸ“œ ê°œë… ì„¤ëª…**

* **í‰ê· ** (`np.mean()`): ëª¨ë“  ì›ì†Œì˜ í‰ê· ê°’
* **ìµœëŒ“ê°’** (`np.max()`): ê°€ì¥ í° ê°’
* **ìµœì†Ÿê°’** (`np.min()`): ê°€ì¥ ì‘ì€ ê°’
* **í•©ê³„** (`np.sum()`): ëª¨ë“  ì›ì†Œì˜ í•©
* **í‘œì¤€í¸ì°¨** (`np.std()`): ë°ì´í„°ê°€ í‰ê· ìœ¼ë¡œë¶€í„° ì–¼ë§ˆë‚˜ í¼ì ¸ìˆëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê°’

#### **ğŸ› ï¸ ì‚¬ìš© ë°©ë²•**
í†µê³„ í•¨ìˆ˜ëŠ” ë°°ì—´ ì „ì²´ì— ì ìš©í•˜ê±°ë‚˜, `axis` ë§¤ê°œë³€ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¹ì • ì¶•(ë°©í–¥)ì„ ê¸°ì¤€ìœ¼ë¡œ ì ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
* `axis=0`: **ì—´(ì„¸ë¡œ)** ë°©í–¥ìœ¼ë¡œ ì—°ì‚°ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.
* `axis=1`: **í–‰(ê°€ë¡œ)** ë°©í–¥ìœ¼ë¡œ ì—°ì‚°ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.

```python
# patient_data ë°°ì—´ ì‚¬ìš©
# ë°°ì—´ ì „ì²´ ì›ì†Œì˜ í‰ê· 
print("ì „ì²´ í‰ê· :", np.mean(patient_data))

# ê° ì—´(íŠ¹ì„±)ì˜ í‰ê·  ê³„ì‚° (axis=0) -> [í‰ê·  í‚¤, í‰ê·  ì²´ì¤‘, í‰ê·  ë‚˜ì´]
column_means = np.mean(patient_data, axis=0)
print("ì—´(íŠ¹ì„±)ë³„ í‰ê· :", column_means)

# ê° í–‰(í™˜ì)ì˜ ìµœëŒ“ê°’ ê³„ì‚° (axis=1)
row_maxes = np.max(patient_data, axis=1)
print("í–‰(í™˜ì)ë³„ ìµœëŒ“ê°’:", row_maxes)
```
