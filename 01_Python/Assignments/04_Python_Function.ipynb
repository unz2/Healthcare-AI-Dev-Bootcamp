## ë¬¸ì œ 5) `**kwargs` â€” ì •ë ¬ëœ 'í‚¤=ê°’' CSV ë¬¸ìì—´

```python
# TODO: join_kv(**info) êµ¬í˜„
# ìš”êµ¬ì‚¬í•­:
# - í‚¤ë¥¼ ì•ŒíŒŒë²³ìˆœìœ¼ë¡œ ì •ë ¬í•œ ë’¤ "í‚¤=ê°’" í˜•íƒœë¡œ ë§Œë“¤ì–´ ì½¤ë§ˆ(,)ë¡œ ì´ì–´ë¶™ì—¬ë¼
# - ì˜ˆ) join_kv(b=2, a=1) -> "a=1,b=2"
# - íŒíŠ¸: sorted(info) ëŠ” ì •ë ¬ëœ í‚¤ ëª©ë¡ì„ ëŒë ¤ì¤€ë‹¤
def join_kv(**info):
    return ",".join(map(lambda k: f"{k}={info[k]}", sorted(info)))

# í…ŒìŠ¤íŠ¸
assert join_kv(b=2, a=1) == "a=1,b=2"
assert join_kv() == ""
print("âœ… ë¬¸ì œ 5 í†µê³¼")
```

## ë¬¸ì œ 15) ì²´ì˜¨ í‰ê·  ê³„ì‚°ê¸°

- ì—¬ëŸ¬ í™˜ìì˜ ì²´ì˜¨ì„ ë°›ì•„ í‰ê· ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
- ì˜µì…˜ `unit` ê°’ì´ `"F"`ë©´ í™”ì”¨(Fahrenheit)ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
- ì•„ë¬´ ë°ì´í„°ë„ ì—†ìœ¼ë©´ `None`ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
- ê²°ê³¼ëŠ” ì†Œìˆ˜ ì²«ì§¸ìë¦¬ê¹Œì§€ ë°˜ì˜¬ë¦¼í•©ë‹ˆë‹¤.


```python
# TODO: avg_temp(*temps, **options) êµ¬í˜„
def avg_temp(*temps, **options):
    """
    í™˜ìë“¤ì˜ ì²´ì˜¨ í‰ê· ì„ ê³„ì‚°
    - *temps: ì—¬ëŸ¬ ì²´ì˜¨ ê°’
    - **options:
        - unit: "C"(ì„­ì”¨, ê¸°ë³¸ê°’) ë˜ëŠ” "F"(í™”ì”¨ ë³€í™˜)
    """
    # TODO: tempsê°€ ë¹„ì–´ìˆìœ¼ë©´ None ë°˜í™˜
    if not temps:
        return None

    # TODO: í‰ê·  ê³„ì‚° (ì´í•© / ê°œìˆ˜)
    avg = sum(temps) / len(temps)

     # TODO: unitì´ "F"ì´ë©´ í™”ì”¨ë¡œ ë³€í™˜
    if options.get('unit') == 'F':
        avg = avg * 1.8 + 32

    # TODO: ì†Œìˆ˜ ì²«ì§¸ ìë¦¬ê¹Œì§€ ë°˜ì˜¬ë¦¼í•˜ì—¬ ë°˜í™˜
    return round(avg,1)


assert avg_temp(36.5, 37.2, 38.0) == 37.2
assert avg_temp(36.5, 37.2, 38.0, unit="F") == 99
assert avg_temp() is None
print("âœ… ë¬¸ì œ 1 í†µê³¼")
```

## ë¬¸ì œ 16) ê²€ì‚¬ ë¹„ìš© ê³„ì‚°ê¸°

- ì—¬ëŸ¬ ê²€ì‚¬ í•­ëª© ë¹„ìš©ì„ ëª¨ë‘ ë”í•©ë‹ˆë‹¤.
- ì˜µì…˜ `insurance=True`ì´ë©´ ì ˆë°˜ ê°€ê²© ì²˜ë¦¬í•©ë‹ˆë‹¤.
- ì˜µì…˜ `discount=0.1`ì´ë©´ 10% ì¶”ê°€ í• ì¸ ì²˜ë¦¬í•©ë‹ˆë‹¤.
- ìµœì¢… ê²°ì œ ê¸ˆì•¡ì„ **ì •ìˆ˜(int)** ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.

```python
# TODO: exam_cost(*costs, **options) êµ¬í˜„
def exam_cost(*costs, **options):
    """
    ê²€ì‚¬ ë¹„ìš© ê³„ì‚°ê¸°
    - *costs: ì—¬ëŸ¬ ê²€ì‚¬ í•­ëª© ë¹„ìš©
    - **options:
        - insurance=True â†’ ì ˆë°˜ ê°€ê²©
        - discount=0.1   â†’ ì¶”ê°€ 10% í• ì¸
    """
    # TODO: ëª¨ë“  ë¹„ìš© í•©ì‚°
    total = sum(costs)

    # TODO: ë³´í—˜ ì ìš© ì‹œ ì ˆë°˜ ê°€ê²© ì²˜ë¦¬
    if options.get('insurance'):
        total *= 0.5

    # TODO: discount ì˜µì…˜ ì ìš©
    if 'discount' in options:
        total *= (1 - options['discount'])

    # TODO: ìµœì¢… ê²°ì œ ê¸ˆì•¡ì„ intë¡œ ë°˜í™˜
    return int(total)


assert exam_cost(10000, 20000, insurance=True) == 15000
assert exam_cost(5000, 5000, discount=0.2) == 8000
assert exam_cost(10000, 20000, insurance=True, discount=0.1) == 13500
print("âœ… ë¬¸ì œ 2 í†µê³¼")
```
