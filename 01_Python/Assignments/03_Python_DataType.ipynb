## ë¬¸ì œ 1) ë¦¬ìŠ¤íŠ¸ ê¸°ë³¸ ë©”ì„œë“œ
- `append, remove, sort, reverse, insert, pop, clear, extend` ë¥¼ ì‚¬ìš©í•´ ìš”êµ¬ì‚¬í•­ì„ ë§Œì¡±í•˜ì„¸ìš”.


```python
fruits = ["apple", "banana", "cherry"]
fruits.__class__  # íƒ€ì… íŒíŠ¸ìš© ë¬´í•´í•œ í‘œí˜„

# 1) ë§¨ ë’¤ì— "orange" ì¶”ê°€
fruits.append('orange')
# 2) "banana" ì‚­ì œ (ì²˜ìŒ í•˜ë‚˜)
fruits.remove('banana')
# 3) ì¸ë±ìŠ¤ 1 ìœ„ì¹˜ì— "grape" ì‚½ì…
fruits.insert(1, 'grape')
# 4) ë” ë§ì€ ê³¼ì¼(kiwi, mango) ì´ì–´ ë¶™ì´ê¸°
more = ['kiwi', 'mango']
fruits.extend(more)
# 5) ì•ŒíŒŒë²³ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
fruits.sort()
# 6) ìˆœì„œ ë’¤ì§‘ê¸°
fruits.reverse()
# 7) ì¸ë±ìŠ¤ 0 ìš”ì†Œë¥¼ êº¼ë‚´ì–´ ë³€ìˆ˜ first_out ì— ì €ì¥
first_out = fruits.pop(0)

assert first_out is not None and isinstance(fruits, list)
assert set(fruits).issubset({"apple","cherry","grape","kiwi","mango","orange"})
print("âœ… ë¬¸ì œ 1 í†µê³¼ | fruits:", fruits, "| first_out:", first_out)
```

## ë¬¸ì œ 2) íŠœí”Œ ë©”ì„œë“œ
- `count`, `index` ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.


```python
t = (1, 2, 3, 2, 2)
cnt = t.count(2)
idx = t.index(3)
assert cnt == 3 and idx == 2
print("âœ… ë¬¸ì œ 2 í†µê³¼")
```

## ë¬¸ì œ 3) ë”•ì…”ë„ˆë¦¬ ë©”ì„œë“œ
- `get, keys, values, items, pop, update, clear` ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.


```python
person = {"name":"Alice", "age":25}

nm = person.get("name")
age_default = person.get("age", 'Unknown')   # ê¸°ë³¸ê°’ í™œìš©
unknown = person.get("address", 'Unknown')

ks = list(person.keys())
vs = list(person.values())
it = list(person.items())

popped_age = person.pop("age")
person.update({"age":26, "city":"Seoul"})

assert nm == "Alice" and age_default == 25 and unknown == "Unknown"
assert set(ks) <= {"name","age","city"}
assert isinstance(it, list)
print("âœ… ë¬¸ì œ 3-1 í†µê³¼ | person:", person)

# ëª¨ë‘ ì‚­ì œ
person.clear()
assert person == {}
print("âœ… ë¬¸ì œ 3-2 í†µê³¼ | cleared person:", person)
```

## ë¬¸ì œ 4) ì§‘í•©(Set) ë©”ì„œë“œ/ì—°ì‚°
- `add, remove, discard, clear, union, intersection, difference` ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.


```python
a = {1,2,3}
b = {3,4,5}

a.add(10)       # add
u = a.union(b)    # union
i = a.intersection(b)    # intersection
d = a.difference(b)    # difference (a-b)

a.remove(2)        # remove 2
a.discard(99)       # discard 99 (ì—ëŸ¬ ì—†ì–´ì•¼ í•¨)

assert u == {1,2,3,4,5,10}
assert i == {3}
assert d.issuperset({1,10}) and 3 not in d
print("âœ… ë¬¸ì œ 4-1 í†µê³¼ | a:", a)

a.clear()         # clear
assert a == set()
print("âœ… ë¬¸ì œ 4-2 í†µê³¼ | cleared a:", a)
```

## ë¬¸ì œ 5) ìë£Œí˜• ë³€í™˜ (Casting)
- ì˜ˆì‹œì— ë§ëŠ” ë³€í™˜ì„ ì±„ìš°ì„¸ìš”.


```python
c1 = list("ABC")            # ['A','B','C']
c2 = tuple([1,2,3])           # (1,2,3)
c3 = set([1,1,2,2])         # {1,2}
c4 = dict([("a",1),("b",2)]) # {'a':1,'b':2}

assert c1 == ['A','B','C'] and c2 == (1,2,3) and c3 == {1,2} and c4 == {'a':1,'b':2}
print("âœ… ë¬¸ì œ 5 í†µê³¼")
```

## ë¬¸ì œ 6) `is` vs `==`
- ë™ì¼ ê°ì²´ ë¹„êµ(`is`)ì™€ ê°’ ë¹„êµ(`==`)ë¥¼ êµ¬ë¶„í•˜ì„¸ìš”.


```python
lst1 = [1,2,3]
lst2 = [1,2,3]
lst3 = lst1

same_value = (lst1 == lst2)
same_object = (lst1 is lst3)

assert same_value is True and same_object is True
print("âœ… ë¬¸ì œ 6 í†µê³¼")
```

## ë¬¸ì œ 7) `in` / `not in`
- í¬í•¨ ì—¬ë¶€ë¥¼ íŒë‹¨í•˜ëŠ” ì½”ë“œë¥¼ ì±„ìš°ì„¸ìš”.


```python
fruits = ["apple","banana","cherry"]
text = "Python Programming"
s = {1,2,3}
d = {"name":"Alice","age":25}

c1 = ("apple" in fruits)
c2 = ("Java" not in text)
c3 = (2 in s)
c4 = ("name" in d)      # dictëŠ” í‚¤ ê¸°ì¤€
c5 = ("Alice" in d.values())

assert [c1,c2,c3,c4,c5] == [True, True, True, True, True]
print("âœ… ë¬¸ì œ 7 í†µê³¼")
```

## ë¬¸ì œ 8) ì‹œí€€ìŠ¤ ìŠ¬ë¼ì´ì‹± ë³µìŠµ
- ë¦¬ìŠ¤íŠ¸ ìŠ¬ë¼ì´ì‹±ì„ ì™„ì„±í•˜ì„¸ìš”: [start:stop:step]


```python
nums = [0,1,2,3,4,5,6]
a = nums[2:5]      # [2,3,4]
b = nums[::2]  # [0,2,4,6]
c = nums[::-1]          # [6,5,4,3,2,1,0]

assert a == [2,3,4] and b == [0,2,4,6] and c == [6,5,4,3,2,1,0]
print("âœ… ë¬¸ì œ 8 í†µê³¼")
```
