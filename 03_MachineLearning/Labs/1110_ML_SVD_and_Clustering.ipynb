```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
img = np.array(Image.open('dog.png').convert('L'))
```

```python
img.shape
```

```python
U, s, VT = np.linalg.svd(img, full_matrices=False)
```

```python
s.shape
```

```python
k_values = [5, 20, 40, 60, 80, 100]
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

# 
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original')
axes[0, 0].axis('off')

#  
axes[0, 1].plot(s[:100])
axes[0, 1].set_title('Singular Values (top 100)')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')

#  
cumsum = np.cumsum(s**2) / np.sum(s**2)
axes[0, 2].plot(cumsum[:100])
axes[0, 2].set_title('Cumulative Energy')
axes[0, 2].axhline(y=0.9, color='r', linestyle='--', label='90%')
axes[0, 2].legend()

for idx, k in enumerate(k_values):
    img_compressed = U[:, :k] @ np.diag(s[:k]) @ VT[:k, :]
    original_size = img.shape[0] * img.shape[1]
    compressed_size = k * (img.shape[0] + img.shape[1] + 1)
    ratio = (1 - compressed_size / original_size) * 100

    # 
    row = 1 + idx // 3
    col = idx % 3
    axes[row, col].imshow(img_compressed, cmap='gray')
    axes[row, col].set_title(f'k={k} (: {ratio:.1f}%)')
    axes[row, col].axis('off')
```

```python
for k in k_values:
    energy = np.sum(s[:k]**2) / np.sum(s**2) * 100
    print(f"k={k}:   {energy:.1f}%")
```

# Mean Shift 

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
#   ( )
X1, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)
X2 = np.random.uniform(-1, 9, (50, 2))  #  
X = np.vstack([X1, X2])
print(f" : {len(X)}")

# Bandwidth  
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)
print(f" bandwidth: {bandwidth:.3f}")

# Mean Shift 
ms = MeanShift(bandwidth=bandwidth)
labels = ms.fit_predict(X)
cluster_centers = ms.cluster_centers_

n_clusters = len(cluster_centers)
print(f"  : {n_clusters}")

#   
for i in range(n_clusters):
    count = np.sum(labels == i)
    print(f" {i}: {count}")
```

```python
# 
plt.figure(figsize=(12, 5))

#  
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], alpha=0.5)
plt.title('Original Data')

# Mean Shift 
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='red', marker='x', s=200, linewidths=3, label='Centers')
plt.title(f'Mean Shift (K={n_clusters})')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#  bandwidth 
bandwidths = [0.5, 1.0, 2.0, 3.0]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for idx, bw in enumerate(bandwidths):
    ms = MeanShift(bandwidth=bw)
    labels = ms.fit_predict(X)
    n_clusters = len(ms.cluster_centers_)
    
    ax = axes[idx // 2, idx % 2]
    ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
    ax.scatter(ms.cluster_centers_[:, 0], ms.cluster_centers_[:, 1],
              c='red', marker='x', s=200, linewidths=3)
    ax.set_title(f'bandwidth={bw} (K={n_clusters})')

plt.tight_layout()
plt.show()
```

# DBSCAN 

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons, make_circles
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

#    
X1, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
X2, _ = make_circles(n_samples=200, noise=0.05, factor=0.5, random_state=0)
X3 = np.random.uniform(-1, 2, (50, 2))  #  

X2[:, 0] += 3  #  
X = np.vstack([X1, X2])
X = np.vstack([X, X3])

# K-distance  eps 
k = 4  # min_samples 
neighbors = NearestNeighbors(n_neighbors=k)
neighbors.fit(X)
distances, indices = neighbors.kneighbors(X)

# k  (  )
k_distances = np.sort(distances[:, k-1])

plt.figure(figsize=(10, 4))
plt.plot(k_distances)
plt.ylabel(f'{k}-th Nearest Neighbor Distance')
plt.xlabel('Points (sorted)')
plt.title('K-distance Graph')
plt.axhline(y=0.3, color='r', linestyle='--', label='eps candidate')
plt.legend()
plt.show()

print(" eps:   (: 0.3)")

# DBSCAN 
dbscan = DBSCAN(eps=0.3, min_samples=4)
labels = dbscan.fit_predict(X)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f" : {n_clusters}")
print(f" : {n_noise}")

#   
for i in range(n_clusters):
    count = np.sum(labels == i)
    print(f" {i}: {count}")

# 
plt.figure(figsize=(12, 5))

# 
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], alpha=0.6)
plt.title('Original Data')

# DBSCAN 
plt.subplot(1, 2, 2)
colors = plt.cm.Spectral(np.linspace(0, 1, n_clusters))
for k, col in zip(range(n_clusters), colors):
    class_member_mask = (labels == k)
    xy = X[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[col], alpha=0.6, 
               label=f'Cluster {k}', s=50)

# 
noise_mask = (labels == -1)
if np.any(noise_mask):
    plt.scatter(X[noise_mask, 0], X[noise_mask, 1], 
               c='black', marker='x', s=50, label='Noise')

plt.title(f'DBSCAN (eps={0.3}, min_samples={4})')
plt.legend()
plt.tight_layout()
plt.show()

#  eps 
eps_values = [0.2, 0.3, 0.5, 0.8]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for idx, eps in enumerate(eps_values):
    dbscan = DBSCAN(eps=eps, min_samples=4)
    labels = dbscan.fit_predict(X)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    
    ax = axes[idx // 2, idx % 2]
    
    #  
    for k in range(n_clusters):
        mask = (labels == k)
        ax.scatter(X[mask, 0], X[mask, 1], alpha=0.6, s=30)
    
    # 
    noise_mask = (labels == -1)
    if np.any(noise_mask):
        ax.scatter(X[noise_mask, 0], X[noise_mask, 1], 
                  c='black', marker='x', s=30)
    
    ax.set_title(f'eps={eps} (K={n_clusters})')

plt.tight_layout()
plt.show()
```
