```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

☝️ seaborn 에서 데이터셋을 불러오면 dataframe으로 불러와지고  
sklearn으로 데이터셋을 불러오면 np.array로 불러와짐

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
df
```

```python
df['species'].value_counts()
```

☝️ scatterplot 은 수치형 + 수치형 데이터로 찍는것이 이상적

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='species')
```

```python
# test 와 train 을 알아서 분리해줌
from sklearn.model_selection import train_test_split
```

```python
# feature
X = df.drop('species', axis=1)
```

```python
# target
Y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
train_df = X_train.copy()
train_df['species'] = y_train
sns.scatterplot(train_df, x='petal length (cm)', y='petal width (cm)', hue='species')
```
